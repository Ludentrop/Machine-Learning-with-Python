```python
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

Требуется сгруппировать наблюдения так, чтобы похожие наблюдения находились в одной группе.

```python
features, _ = make_blobs(
    n_samples=50,
    n_features=2,
    centers=3,
    random_state=1
)

data = pd.DataFrame(features, columns=['feature_1', 'feature_2'])

# The clusterer creating using the k-means method
clusterer = KMeans(n_clusters=3, random_state=1, n_init=10)

# Fitting the clusterer
clusterer.fit(features)

data['group'] = clusterer.predict(features)

data.head(5)
```
