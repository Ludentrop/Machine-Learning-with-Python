```python
import numpy as np
from sklearn import preprocessing
```

To scale features massive use *MinMaxScaler* [x'i = (xi - min(x)) / (max(x)  - min(x))] method or Rescaling (min-max normalization). Scaling is a common preprocessing task in ML.  There are two ways od scaling: use *fit* method to find out min and max feature value, then use *transform* method to scale the feature; or use *fit_transform* to make all the operations at once. There is no difference use 1 or 2 way, it's about comfort.

```python
feature = np.array([
    [-500.5],
    [-100.1],
    [0],
    [100.1],
    [900.9]
])


# Creating a scaler
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0, 1))

# Scaling a feature
scaled_feature = minmax_scale.fit_transform(feature)

scaled_feature
```

```python
import matplotlib.pyplot as plt


f, a = plt.subplots(1, 2)

a[0].plot(feature, range(len(feature)))
a[0].set_title('Raw feature')
a[1].plot(scaled_feature, range(len(scaled_feature)))
a[1].set_title('Scaled feature')
plt.show()
```
