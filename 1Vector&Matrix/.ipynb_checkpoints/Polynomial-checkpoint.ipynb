```python
import numpy as np
from sklearn import preprocessing
```

Полиномиальные признаки часто создаются, когда мы предполагаем, что существует нелинейная связь между признаками и целью. Параметр degree определяет максимальный порядок полинома: x1, x2, x1^2, x2^2.

```python
features = np.array([
    [2, 3],
    [2, 3],
    [2, 3]
])
```

```python
polynomial_interaction = preprocessing.PolynomialFeatures(degree=2, include_bias=False)

# Creating polynomial features
polynomial_interaction.fit_transform(features)
```

Мы можем ограничить создаваемые признаки только признаками взаимодействия, установив interaction_only значение True.

```python
polynomial_interaction = preprocessing.PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)

polynomial_interaction.fit_transform(features)
```

Кроме того, нередко мы сталкиваемся с ситуациями, когда эффект одного признака зависит от еще одного признака. Влияние каждого признака на цель зависит друг от друга. Мы можем кодировать эту связь, включив взаимодействующий признак, который является произведением отдельных признаков.
