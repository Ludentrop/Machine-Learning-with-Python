```python
import numpy as np
from sklearn import preprocessing
```

```python
features = np.array([
    [2, 3],
    [2, 3],
    [2, 3]
])
```

```python
add_ten = lambda x: x + 10

ten_transformer = preprocessing.FunctionTransformer(add_ten)

ten_transformer.transform(features)
```

```python
import pandas as pd


df = pd.DataFrame(features, columns=['feature_1', 'feature_2'])

df.apply(add_ten)
```
