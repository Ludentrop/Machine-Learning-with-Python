Дан категориальный признак, содержащий пропущенные значения, которые требуется заменить предсказанными значениями.

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
# The feature matrix with the categorial features
X = np.array([[0, 2.10, 1.45],
              [1, 1.18, 1.33],
              [0, 1.22, 1.27],
              [1, -0.21, -1.19]])

# The feature matrix with missing values in the categorial feature
X_with_nan = np.array([
    [np.nan, 0.87, 1.31],
    [np.nan, -0.67, -1.22]
])
```

```python
# Train KNN
clf = KNeighborsClassifier(3, weights='distance')
trained_model = clf.fit(X[:, 1:], X[:, 0])
```

```python
# Predict missing values class
imputed_values = trained_model.predict(X_with_nan[:, 1:])
```

```python
# Merge predicted class column with other features
X_with_imputed = np.hstack((imputed_values.reshape(-1, 1), X_with_nan[:, 1:]))
```

```python
# Merge two feature matrices
np.vstack((X_with_imputed, X))
```

Альтернативным решением является заполнение пропущенных значений наиболее частыми значениями признаков

```python
from sklearn.impute import SimpleImputer
```

```python
X_complete = np.vstack((X_with_nan, X))
imputer = SimpleImputer(strategy='most_frequent')
imputer.fit_transform(X_complete)
```
