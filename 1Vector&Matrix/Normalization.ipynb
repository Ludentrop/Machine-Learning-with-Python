```python
import numpy as np
from sklearn import preprocessing
```

Многие методы шкалирования (например, минимаксное шкалирование и стандартизация) работают с признаками; однако мы также можем шкалировать отдельные наблюдения. Класс Normaiizer шкалирует значения в отдельных наблюдениях, приводя их к единичной норме (сумма их длин равна 1). Этот тип шкалирования часто используют, когда имеется много эквивалентных признаков (например, в классификации текста, когда каждое слово или группа и-слов является признаком). Класс Normaiizer предоставляет три варианта нормы, при этом евклидова норма (нередко именуемая L^2 -нормой) является аргументом [|x|^2 = sqrt(x1^2 + x2^2 + ... + xn^2)]

```python
features = np.array([
    [.5, .5],
    [1.1, 3.4],
    [1.5, 20.2],
    [1.63, 34.4],
    [10.9, 3.3]
])


normalizer = preprocessing.Normalizer(norm='l2')

features_l2_norm = normalizer.transform(features)

features_l2_norm
```

В качестве альтернативы можно указать манхэттенскую норму ( Z^1) [|x|^1 = SUMi=1,n |xi|]

```python
normalizer = preprocessing.Normalizer(norm='l1')

features_l1_norm = normalizer.transform(features)

features_l1_norm
```

```python
import matplotlib.pyplot as plt


f, a = plt.subplots(1, 3)

a[0].plot(features)
a[0].set_title('Raw feature')
a[1].plot(features_l1_norm)
a[1].set_title('features_l1_norm')
a[2].plot(features_l2_norm)
a[2].set_title('features_l2_norm')
plt.show()
```
