```python
import pandas as pd
import numpy as np
```

There are three strategies of the outliers processing:

1.   Drop the outliers;


```python
houses = pd.DataFrame()
houses['Price'] = [534433, 392333, 293222, 4322032]
houses['Bethrooms'] = [2, 3.5, 2, 116]
houses['Square foots'] = [1500, 2500, 1500, 48000]

houses[houses['Bethrooms'] < 20]
```




---


2.   Mark outliers and include them as features



```python
houses['Outlier'] = np.where(houses['Bethrooms'] < 20, 0, 1)

houses
```


---

3.  Weaken the outlier effect

```python
houses['log_sqr_ft'] = [np.log(x) for x in houses['Square foots']]

houses
```
