```python
import numpy as np
from sklearn import preprocessing
```

It is need to transform a feature so that it has an average 0 and standard deviation 1. We need to use *StandardScaler* class of scikit-learn. This is the MinMax alternative, in this case features are standardly distributed. [x'i = (xi - x) / sigma]. The transformed feature is the number of standard deviations by which the initial value is separated from the average value of the feature (Z-score Normalization)

```python
x = np.array([
    [-1000.1],
    [-200.2],
    [500.5],
    [600.6],
    [9000.9]
])


scaler = preprocessing.StandardScaler()

standardized = scaler.fit_transform(x)

[i for i in standardized]
```

```python
print(round(standardized.mean()))
print(standardized.std())
```

If our data has significant outliers, this can negatively affect standardization, affecting the mean value and variance of the feature. In this case, instead, it is often useful to scale the feature using the median and the quarterly span. In scikit-learn, we do this using the *Robustscaier* class

```python
robust_scaler = preprocessing.RobustScaler()

robusted = robust_scaler.fit_transform(x)

robusted
```

```python
import matplotlib.pyplot as plt


f, a = plt.subplots(1, 3)

a[0].plot(x, range(len(x)))
a[0].set_title('Raw feature')
a[1].plot(standardized, range(len(standardized)))
a[1].set_title('Standardized feature')
a[2].plot(robusted, range(len(robusted)))
a[2].set_title('Robusted')
plt.show()
```
