```python
from sklearn.datasets import make_regression, make_classification, make_blobs
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import mplcyberpunk

plt.style.use('cyberpunk')
```

```python
digits = datasets.load_digits()
features = digits.data
target = digits.target
print(features[0])
```

## Creating simulated data for the **regression problem**

```python
features, target, coefficients = make_regression(
    n_samples=100,
    n_features=3,
    n_informative=3,
    n_targets=1,
    noise=0.0,
    coef=True,
    random_state=1
)

print(features[:3])
print()
print(target[:3])
```

## Creating simulated data for the **classification problem**

```python
features, target = make_classification(
    n_samples=100,
    n_features=3,
    n_informative=3,
    n_redundant=0,
    n_classes=2,
    weights=[.25, .75],
    random_state=1
)

print(features[:3])
print()
print(target[:3])
```

## Creating simulated data for the clustering methods

```python
features, target = make_blobs(
    n_samples=100,
    n_features=2,
    centers=3,
    cluster_std=.5,
    shuffle=True,
    random_state=1
)

print(features[:3])
print(target[:3])
```

```python
plt.scatter(features[:, 0], features[:, 1], c=target)
```

```python
sns.scatterplot(x=features[:, 0], y=features[:, 1], hue=target)
```
